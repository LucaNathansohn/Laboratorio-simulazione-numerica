### <span style="color:teal"> Esercizio 10.1 </span>

In questo esercizio ho esteso il codice sviluppato nella scorsa esercitazione rendendolo parallelizzabile con MPI (Message Passing Interface).

La parallelizzazione consiste in una ricerca **indipendente** della miglior soluzione da parte di 11 core.

Si creano 11 "continenti" che sviluppano in autonomia la loro soluzione ottimale utilizzando il GA, ogni $N_{migr}$  generazioni, due continenti scelti casualmente si scambiano il loro miglior individuo.

Così facendo, ogni continente esplora a fondo una zona diversa dello spazio delle soluzioni possibili, permettendo di mantenere diversità genetica tra un continente e l'altro. Questo aumenta la qualità della soluzione finale. Infatti, grazie al meccanismo delle migrazioni, verrà selezionata la miglior soluzione tra diversi minimi locali del problema.

#### Modifiche al codice

La struttura di base è rimasta quella dell'esercitazione 9, le modifiche principali sono state:
- aggiungere il "rank" agli argomenti del costruttore di "Popolazione" in modo da poter inizializzare simulazioni indipendenti per ogni core.
- Nel main, nel ciclo sulle generazioni ho introdotto un meccanismo di migrazione che segue la logica decritta sopra e si attiva ogni $N_{migr}=70$ generazioni. Per implementarlo ho fatto uso di MPI_send, MPI_Recv e MPI_Wait per gestire la comunicazione tra processori.

### <span style="color:teal"> Esercizio 10.2 </span>

I grafici che seguono riportano i risultati dell'applicazione dell'algoritmo parallelizzato a un TSP tra i 110 capoluoghi di provincia italiani

```python
import matplotlib.pyplot as plt
import glob
import numpy as np
import os

# === Imposta il range da plottare (in generazioni) ===
start_gen = 300     # incluso
end_gen = 4000      # incluso

# === Carica tutti i file ===
file_list = sorted(glob.glob("ave_*.dat"))

plt.figure(figsize=(10, 6))

for filename in file_list:
    data = np.loadtxt(filename)
    
    # Indici Python: start incluso, end escluso
    y = data[start_gen:end_gen, 1]  # Colonna della lunghezza best
    x = np.arange(start_gen + 1, end_gen + 1)  # Generazioni (da 301 a 4000)

    # Etichetta pulita: es. da 'ave_1.dat' → 'loss best 1'
    base_name = os.path.splitext(os.path.basename(filename))[0]  # es. 'ave_1'
    label_name = base_name.replace('ave_', '$L^{(1)}$ del miglior inidividuo per il core ')

    plt.plot(x, y, label=label_name)

# === Aggiungi linea orizzontale a y = 64.8528 con etichetta sull’asse ===
target_y = 64.8528
plt.axhline(y=target_y, color='red', linestyle='--', linewidth=1)
plt.text(x=start_gen + 10, y=target_y + 0.1, s=f"{target_y:.4f}", color='red', fontsize=10)

# === Impostazioni del grafico ===
plt.xlabel("Numero generazione")
plt.ylabel("Lunghezza best")
plt.title(f"Andamento del fitness per i diversi core (generazioni {start_gen + 1}–{end_gen})")
plt.legend(fontsize='small', loc='best')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Il grafico sopra mostra l’andamento del fitness $L^{(1)}$ del miglior percorso in funzione del numero di generazioni dell’algoritmo genetico, per ogni processore.

Possiamo vedere che, alla fine della simulazione, tutti i processori sono arrivati alla stessa soluzione finale.

L'andamento mostrato è ristretto all'intervallo di generazioni 300-4000, in questo modo, oltre al solito grafico decrescente che mostra la convergenza al valore finale, si osservano bene i punti di salto della funzione, dovuti alla migrazione di percorsi migliori provenienti da altri core.

```python
import matplotlib.pyplot as plt

def leggi_citta(file_citta):
    coords = {}
    with open(file_citta, 'r') as f:
        for i, riga in enumerate(f):
            parti = riga.strip().split()
            if len(parti) >= 2:
                idx = i + 1
                x = float(parti[0])
                y = float(parti[1])
                coords[idx] = (x, y)
    return coords

def leggi_ultimo_percorso(file_best):
    with open(file_best, 'r') as f:
        righe = f.readlines()
    ultima_riga = righe[-5]
    percorso = list(map(int, ultima_riga.strip().split()))
    return percorso

def plot_percorso_subplot(ax, percorso, coords, titolo):
    x = [coords[c][0] for c in percorso]
    y = [coords[c][1] for c in percorso]
    x.append(coords[percorso[0]][0])
    y.append(coords[percorso[0]][1])

    ax.plot(x, y, marker='o', linestyle='-', color='teal')  # Teal invece di blu
    for city in percorso:
        ax.text(coords[city][0], coords[city][1], str(city), fontsize=6, ha='right')
    ax.set_title(titolo)
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.grid(True)

def leggi_loss(file_ave):
    media = []
    migliore = []
    with open(file_ave, 'r') as f:
        for riga in f:
            parti = riga.strip().split()
            if len(parti) >= 2:
                media.append(float(parti[0]))
                migliore.append(float(parti[1]))
    return media, migliore

def plot_loss_subplot(ax, media, migliore, titolo, max_generazioni=500):
    # Tronca a max_generazioni se necessario
    media = media[:max_generazioni]
    migliore = migliore[:max_generazioni]
    generazioni = list(range(len(media)))

    ax.plot(generazioni, migliore, label=r"$L^{(1)}$ best", color='navy')
    ax.plot(generazioni, media, label=r"$\langle L^{(1)} \rangle$", color='teal')
    ax.set_title(titolo)
    ax.set_xlabel("Generazione")
    ax.set_ylabel("$L^{(1)}$")
    ax.grid(True)
    ax.legend()

if __name__ == "__main__":
    # File paths
    citta_file = "cap_prov_ita.dat"
    best_file_1 = "best_5.dat"
    ave_file_1 = "ave_5.dat"
    best_file_2 = "OUTPUT_1_CORE/best.dat"
    ave_file_2 = "OUTPUT_1_CORE/ave.dat"

    # Lettura dati
    coords = leggi_citta(citta_file)
    percorso1 = leggi_ultimo_percorso(best_file_1)
    percorso2 = leggi_ultimo_percorso(best_file_2)
    media1, migliore1 = leggi_loss(ave_file_1)
    media2, migliore2 = leggi_loss(ave_file_2)

    # Plot
    fig, axs = plt.subplots(2, 2, figsize=(16, 10))

    plot_percorso_subplot(axs[0, 0], percorso1, coords, "Miglior percorso con 11 CORE")
    plot_loss_subplot(axs[0, 1], media1, migliore1, "Andamento del fitness per le prime 500 generazioni (11 CORE)", max_generazioni=500)

    plot_percorso_subplot(axs[1, 0], percorso2, coords, "Miglior percorso con 1 CORE")
    plot_loss_subplot(axs[1, 1], media2, migliore2, "Andamento del fitness per le prime 500 generazioni (1 CORE)", max_generazioni=500)

    plt.tight_layout()
    plt.show()
```

Nei grafici sopra vengono confrontati i risultati ottenuti con 11 processori (1000 individui e 4000 generazioni per ogni simulazione indipendente) con quelli derivati da una singola simulazione con un solo core, ma con lo stesso numero di individui degli 11 core messi insieme (11000 individui in uno stesso continente con sempre 4000 generazioni). I valori finali di $L^{(1)}$ per il miglior individuo sono stati:
   - 66.322 per la simulazione con un singolo processore
   - 64.8528 per la simulazione parallelizzata

La differenza tra i risultati è significativa, sintomo di un codice altamente parallelizzabile.

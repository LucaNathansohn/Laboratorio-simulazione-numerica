```python
import matplotlib.pyplot as plt
import numpy as np

# Lista dei valori di N da analizzare
N_values = [1, 2, 10, 100]

# Creazione di una figura con pi subplot
fig, axes = plt.subplots(len(N_values), 3, figsize=(15, 4 * len(N_values)))

for i, N in enumerate(N_values):
    # Carica il file corrispondente a N
    filename = f"histo_N{N}.dat"
    data = np.loadtxt(filename)

    # Estrai le distribuzioni
    dice = data[:, 0]
    exp = data[:, 1]
    lorentz = data[:, 2]

    # Istogramma per la distribuzione uniforme (dadi)
    axes[i, 0].hist(dice, bins=50, color='skyblue', edgecolor='black', density=True)
    axes[i, 0].set_title(f'Dice Distribution (N={N})')
    axes[i, 0].set_xlabel('Value')
    axes[i, 0].set_ylabel('Frequency')
    axes[i, 0].grid(True)

    # Istogramma per la distribuzione esponenziale
    axes[i, 1].hist(exp, bins=50, color='orange', edgecolor='black', density=True)
    axes[i, 1].set_title(f'Exponential Distribution (N={N})')
    axes[i, 1].set_xlabel('Value')
    axes[i, 1].set_ylabel('Frequency')
    axes[i, 1].grid(True)

    # Istogramma per la distribuzione Lorentziana
    axes[i, 2].hist(lorentz, bins=50, color='red', edgecolor='black', density=True)
    axes[i, 2].set_title(f'Lorentzian Distribution (N={N})')
    axes[i, 2].set_xlabel('Value')
    axes[i, 2].set_ylabel('Frequency')
    axes[i, 2].grid(True)

# Adatta lo spazio tra i subplot
plt.tight_layout()
plt.show()
```

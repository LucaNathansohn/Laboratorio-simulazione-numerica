```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

M=100000              # Total number of throws
N=100                 # Number of blocks

L=int(M/N)            # Number of throws in each block, please use for M a multiple of N
x = np.arange(N)      # [0,1,2,...,N-1]

data_medie = np.loadtxt("medie.dat")

medie = data_medie[:,0]
errore_medie = data_medie[:,1]

x*=L # Number of throws = block * (Number of throws in each block)
plt.errorbar(x,medie-0.5,yerr=errore_medie)
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

```python

import numpy as np
x = np.arange(N)      # [0,1,2,...,N-1]

data_varianze = np.loadtxt("varianza.dat")

varianze = data_varianze[:,0]
errore_varianze = data_varianze[:,1]

x*=L # Number of throws = block * (Number of throws in each block)
plt.errorbar(x,varianze-1/12,yerr=errore_varianze)
plt.xlabel('#throws')
plt.ylabel('<(r-0.5)^2>-1/12')
plt.grid(True)
plt.show()
```

```python
import numpy as np
data_chi2 = np.loadtxt("Chi2.dat")

mean_chi2 = np.mean(data_chi2)
std_chi2 = np.std(data_chi2)
print(f"Media di Chi2: {mean_chi2}")
print(f"Deviazione standard di Chi2: {std_chi2}")

plt.hist(data_chi2, bins=20, color='skyblue', edgecolor='black')  # Creiamo l'istogramma
plt.title('Distribuzione di Chi2')  # Titolo del grafico
plt.xlabel('Chi2')  # Etichetta sull'asse x
plt.ylabel('Frequenza')  # Etichetta sull'asse y
plt.grid(True)  # Aggiungi una griglia
plt.show()  # Mostra il grafico
```

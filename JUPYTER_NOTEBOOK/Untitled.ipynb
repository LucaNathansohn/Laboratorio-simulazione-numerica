```python


import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.random.rand(100).astype(np.float32)
y = 3 * x + 2 + np.random.normal(0, 0.1, size=x.shape).astype(np.float32)

# Uso dell'Input esplicito per evitare il warning su input_shape
model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(units=1)
])

model.compile(optimizer='sgd', loss='mean_squared_error')
history = model.fit(x, y, epochs=100, verbose=0)

weights = model.get_weights()
print(f"Peso appreso: {weights[0][0][0]:.2f}, Bias appreso: {weights[1][0]:.2f}")

y_pred = model.predict(x)
plt.scatter(x, y, label='Dati reali')
plt.plot(x, y_pred, color='red', label='Predizione')
plt.legend()
plt.title("TensorFlow: Regressione Lineare")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
